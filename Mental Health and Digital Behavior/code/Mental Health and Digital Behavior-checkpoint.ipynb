```python
import pandas as pd
#Dosyayı okutalım
df=pd.read_csv("mental_health_digital_behavior_data.csv")
df
```

```python
#Veri Tipi
print(df.info())
```

```python
#Eksik değerleri görelim
print("\n Eksik Veri sayıları")
print(df.isnull().sum())
```

Eksik Değer olmadığını gözlemliyoruz :)

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Sadece Sayısal Sütunları alıyoruz
numeric_df=df.select_dtypes(include="number")
#Korelasyon Matrisi
corr_matrix=numeric_df.corr()
#Isı Haritası
plt.figure(figsize=(16,12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Sayısal Değişkenler Arası Korelasyon Isı Haritası")
plt.show()           
           
```

```python
#Uyku saatini Kategorilere ayırıyoruz
bins=[0,6,8,12]
labels=["<6 saat","6-8 saat", ">8 saat"]
df["sleep_category"]= pd.cut(df["sleep_hours"], bins=bins, labels=labels, include_lowest=True)
#2. Her Kategoriye göre dijital iyi oluş skorlarını güncelleştir(boxplot)
plt.figure(figsize=(8,6))
sns.boxplot(x="sleep_category", y="digital_wellbeing_score", data=df, hue=None, palette="Set2", dodge=False)
plt.title("Uyku Kategorile Göre Dijital İyi Oluş Skoru")
plt.xlabel("Uyku Süresi")
plt.ylabel("Dijital İyi Oluş Skoru")
plt.grid(True)
plt.show()
```

```python
#Anksiyete Seviyelerini Kategorilere Ayıralım
bins_anxiety=[0,4,7,10]
labels_anxiety=["Düşük","Orta","Yüksek"]
df["anxiety_category"]=pd.cut(df["anxiety_level"],bins=bins_anxiety, labels=labels_anxiety, include_lowest=True)

#Sosyal Medya Sürelerini Kategorilere Ayırıyoruz
bins_social=[0,90,180,1000]
labels_social=["Az","Orta","Çok"]
df["social_media_category"]=pd.cut(df["social_media_time_min"],bins=bins_social, labels=labels_social, include_lowest=True)

#Anksiyete kategorilerine göre odak skoru
plt.figure(figsize=(8,6))
sns.boxplot(x="anxiety_category", y="focus_score",data=df, palette="Set3")
plt.title("Anksiyete Seviyelerine Göre Odaklanma Skoru ")
plt.xlabel("Anksiyete Seviyesi")
plt.ylabel("Odaklanma Skoru")
plt.grid(True)
plt.show()

#Sosyal Medya Kullanımına Göre Mood Skoru
plt.figure(figsize=(8,6))
sns.boxplot(x="social_media_category", y="mood_score",data=df,palette="Set1")
plt.title("Sosyal Medya Kullanımı Gruplarına Göre Ruh Hali(Modo) Skoru ")
plt.xlabel("Sosyal Medya Kullanım Düzeyi")
plt.ylabel("Mood Skoru")
plt.grid(True)
plt.show()


```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
#Atama yapıyoruz
x=df["sleep_hours"]
y=df["anxiety_level"]
#Scatter plot+ regresyon çizgisi 
plt.figure(figsize=(8,6))
sns.regplot(x=x,y=y,ci=95,line_kws={"color":"red"})
plt.xlabel("Sleep Hours")
plt.ylabel("Anxiety Level")
plt.title("Uyku Süresi ile Anksiyete Düzeyi Arasındaki İlişki")
plt.grid(True)
plt.show()
```

```python
#Lineer Regresyon modeli oluştur(statsmodels)
X=sm.add_constant(x) #Sabit terim eklemek için
model=sm.OLS(y,X).fit()
model_summary=model.summary()
model_summary
```

Bu durum, başka değişkenlerin etkisini kontrol etmeden (örneğin çoklu regresyon) uyku süresinin tek başına belirleyici olmadığını gösteriyor olabilir.
